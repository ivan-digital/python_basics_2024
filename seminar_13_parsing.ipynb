Сделаем простой (некрасивый) парсер для ссылок на курсы КНАД

```python
for elem in soup('td'):
    if 'КНАД' in elem.text:
        for child in elem.children:
            if re.findall(r'([\w\s()]+)КНАД', child.text):
                for a in child('a'):
                    title = a.get('title', 'No title')
                    print(f"{title}: http://wiki.cs.hse.ru{a['href']}")
```

## Search GitHub API

Проблема в html парсинге в том, что код страницы на фронте может постоянно меняться (и придется постоянно подправлять ваш парсер). Гораздо удобнее, если у сайта есть API -- программный интерфейс для работы с ним. Например, github предоставляет такой

Напишем программу, которая будет делать поиски по репозиториям на гитхабе и сохранять некоторые данные о владельцах репозиториев

```python
%%writefile github_sequential.py

import time
from typing import Callable
import requests

import os

BASE_DIR = 'github_downloads'
URL = 'https://api.github.com/search/repositories'
KEYWORDS = ['search', 'torch', 'tensorflow',
            'requests', 'alice', 'data', 'deep', 'ML']

def get_repos_data(query: str) -> dict:
    response = requests.get(
        URL,
        params={'q': f'{query}+language:python'},
    )
    if response.ok:
        return response.json()
    else:
        return {}


def get_owners_ids(resp: dict) -> list:
    owners = []
    for item in resp.get('items', []):
        owners.append(f"({item['owner']['login']}: {item['owner']['html_url']}, {item['owner']['id']})")
    return owners


def save_stats(ids: list, query_name: str):
    print(f'Saving {len(ids)} stats for {query_name}')
    os.makedirs(BASE_DIR, exist_ok=True)
    with open(os.path.join(BASE_DIR, query_name), 'w') as f:
        f.write(str(ids))


def get_many_stats(queries: list) -> int:
    for query in queries:
        data = get_repos_data(query)
        repo_owners = get_owners_ids(data)
        save_stats(repo_owners, query)
    return len(queries)

def main(downloader: Callable[[list], dict]):
    t_start = time.perf_counter()
    count = downloader(KEYWORDS)
    elapsed = time.perf_counter() - t_start
    print(f'\n{count} searches done in {elapsed:.2f}s')

if __name__ == '__main__':
    main(get_many_stats)
```

Асинхронная версия будет работать быстрее

```python
%%writefile github_async.py

import asyncio
import aiohttp
from aiohttp import ClientSession

from github_sequential import URL, main, save_stats, get_owners_ids

async def get_repos_data(session: ClientSession, query: str) -> dict:
    query_url = f'{URL}?q={query}+language:python'
    async with session.get(query_url) as response:
        return await response.json()


async def save_one(session: ClientSession, query: str):
    data = await get_repos_data(session, query)
    ids = get_owners_ids(data)
    save_stats(ids, query)

def save_many(queries: list) -> int:
    return asyncio.run(supervisor(queries))

async def supervisor(queries: list) -> int:
    async with ClientSession() as session:
        to_do = [save_one(session, q) for q in queries]
        res = await asyncio.gather(*to_do)

    return len(res)

if __name__ == '__main__':
    main(save_many)
```

```python
# Для локального запуска на маке, нужно скачать драйвар предварительно

from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

options = Options()
options.add_argument("--headless")

geckodriver_path = "/Users/ivan/PycharmProjects/seminar_bot/geckodriver"

service = Service(executable_path=geckodriver_path)
driver = webdriver.Firefox(service=service, options=options)


try:
    url = "https://duckduckgo.com"
    driver.get(url)

    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "title")))

    print("Page Title:", driver.title)


finally:
    # Close the browser
    driver.quit()
```
